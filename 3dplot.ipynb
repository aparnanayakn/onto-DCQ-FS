```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

matplotlib.rc('font', **font)
```

```python
fig = plt.figure(figsize=(16,12))
ax = plt.axes(projection='3d')

evalMeasures = ['','Divergence', 'Accuracy', 'Consistency', 'Information', 'Dependence', 'Distance']
goSucc = ['', ' ', 'Forward', 'Backward', 'Compound','Weighting', 'Random']
searchOrga = ['Exponential', 'Sequential', 'Random']
x=np.arange(len(evalMeasures)+1)
ax.set(xticks= x, xticklabels = evalMeasures)
y=np.arange(len(goSucc)+1)
ax.set(yticks= y, yticklabels = goSucc)
z=np.arange(len(searchOrga)+1)


ax.set(zticks= z, zticklabels = searchOrga)
plt.xlabel('Search organization',  fontsize=20, labelpad=20)
plt.ylabel('Generation of successors', fontsize=20, labelpad=22)
ax.set_zlabel('Evaluation measure',  fontsize=20, labelpad=30);

fsa = ['algo1', 'algo2', 'algo3', 'algo4','algo5','algo6','algo7','algo8','algo9']

x=[1,2,3,4,5,6,1,2,3]
y=[6,5,4,3,2,1,1,2,3]
z=[1,2,3,1,2,3,1,2,3]
ax.scatter(x, y, z)

zs_l = np.asarray([[i, -1] for i in z])

for i, txt in enumerate(fsa):
    ax.text(x[i], y[i], z[i], txt)

cl = [cm.get_cmap('viridis_r')(i/len(z)) for i in range(len(z))]
 
for i, p in enumerate(zs_l):
    ax.plot(xs=[x[i]]*2, ys=[y[i]]*2, zs=zs_l[i], markersize=0, lw=3, c=cl[i])

for label in (ax.get_xticklabels() + ax.get_yticklabels() + ax.get_zticklabels()):
	label.set_fontsize(16)

plt.savefig("qwe.pdf")
```

```python
import matplotlib.pyplot as plt


ax = plt.figure().add_subplot(projection='3d')

# Demo 1: zdir
zdirs = (None, 'x', 'y', 'z', (1, 1, 0), (1, 1, 1))
xs = (1, 4, 4, 9, 4, 1)
ys = (2, 5, 8, 10, 1, 2)
zs = (10, 3, 8, 9, 1, 8)

for zdir, x, y, z in zip(zdirs, xs, ys, zs):
    label = '(%d, %d, %d), dir=%s' % (x, y, z, zdir)
    ax.text(x, y, z, label, zdir)

# Demo 2: color
ax.text(9, 0, 0, "red", color='red')

# Demo 3: text2D
# Placement 0, 0 would be the bottom left, 1, 1 would be the top right.
ax.text2D(0.05, 0.95, "2D Text", transform=ax.transAxes)

# Tweaking display region and labels
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_zlim(0, 10)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```
